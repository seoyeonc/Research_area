`-` 맥락

- 신용카드 회사는 고객이 구매하지 않은 품목에 대해 청구되지 않도록, 신용카드 사기 거래를 인식하는 것이 중요합니다.

`-` 내용

- 이 데이터셋은 2013년 9월, 유럽 카드 소유자가 만든 거래를 포함합니다.
- 2일 동안의 거래가 포함되어 있으며, 총 284,807건 중 492건(0.172%)이 사기로 분류된 불균형 데이터셋입니다.
- 입력 변수는 PCA 변환을 통해 얻은 수치형 데이터로, 원래의 특성과 배경 정보는 제공되지 않습니다.
- Time: 첫 거래로부터 각 거래까지 경과 시간(초).
- Amount: 거래 금액. 비용 민감 학습에 사용할 수 있음.
- Class: 반응 변수. 1(사기), 0(정상 거래).

`-`권장 사항

- 데이터 불균형으로 인해 **정확도는 Precision-Recall Curve(AUPRC)**로 측정하는 것이 적합합니다.
- 혼동행렬 기반 정확도는 유의미하지 않음.
